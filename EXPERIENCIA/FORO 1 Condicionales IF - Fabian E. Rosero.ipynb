Crea un programa en Python que utilice condicionales anidados para determinar si la persona puede conducir el tipo de vehículo elegido y muestre un mensaje apropiado.


```python
tiene_licencia=True
edad=18
vehiculo="automóvil"

if tiene_licencia:
    if edad<21:
        if vehiculo=="motocicleta":
            print("Puede conducir motocicleta")
        else:
            print("Usted solo puede conducir motocicleta")
    else:
        print(f"Usted puede conducir {vehiculo}")
else:
    print ("Usted No tiene licencia, No puede conducir ningún vehículo")
            
```

### ***3.***

Determinar si un número ingresado por teclado es primo

```python
prim=int(input("Favor,Ingrese un número"))
# contadores para controlar el numero de veces que encuentra si 
# el modulo es 0 por tanto no es primo, se limita parea que no haga mas recorridos
# en caso el numero sea muy grande, asi mismo se controla cuando no encuentra modulo 0 entonces es primo.
cont_no=0
cont_si=0
if prim<=1:
    print("Un número primo es un número natural mayor que 1")
elif prim==2:
    print(f"El numero {prim} es primo")
else:
    # ciclo for para recorrer la lista con numero limitados por el 
    # numero ingresado, cada ciclo guarda en la variable mod el resultado de 
    # hacer modulo con i es decir los elementos de la lista
    for i in range(2,prim):
        mod=prim%i
        #cada ciclo la variable cont_si se incrementa
        cont_si+=1
        #compara si el modulo es 0 incrementa el cont_no (no primo)
        if mod==0:
            cont_no+=1
        if cont_no>4: #rompe ciclo si ya ha encontrado mas de 4 veces el 0
            break
        if cont_si>20: #rompe ciclo cuando no hay 0 entonces cont_si(si primo) hasta 21 veces
            break
    #compara si el contador es diferente de 0 es porque encontró modulos 0 (numero NO primo) 
    if cont_no!=0:
        print (f"Numero {prim} No es Primo")
    else:
        print(f"El número {prim} es Primo") #imprime en caso contrario osea es primo     
```
