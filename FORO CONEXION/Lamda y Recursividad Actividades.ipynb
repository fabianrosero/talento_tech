```python
'''''CALCULADORA BASICA USANDO FUNCIONES LAMDA
EJERCICIOS DE FACTORIAL
EJERCICIO FIBONACCIO RECURSIVO'
CREACION LISTA PANDA

EJERCICIOS PROPIOS PARA CREAR UN MENU CON WHILE VALIDAR OPCION, OTRO VALIDAR TILDES'''


import math
n1 = float(input("Enter the 1st number: "))
n2 = float(input("Enter the 2nd number: "))

suma = lambda a,b: a+b
resultado = suma(n1, n2)
print(f"The sum is: {resultado}")

resta = lambda a,b: a-b
resultado = resta(n1, n2)
print(f"The substraction is: {resultado}")

multiplicacion = lambda a,b: a*b
resultado = multiplicacion(n1, n2)
print(f"The multiplication is: {resultado}")

division = lambda a,b: a/b
resultado = division(n1, n2) if n2!=0 else "You cannot divide by zero"
print(f"The division is: {resultado}")

potencia = lambda base,exponente: math.pow(base,exponente)# base**exponente
resultado = potencia(n1, n2)
print(f"The Pow base {n1} exp {n2} is: {resultado}")

raiz = lambda base,exponente: math.pow(base,1/exponente)
resultado = raiz(n1, n2)
print(f"The Root {n2} of {n1} is: {resultado}")

logaritmo = lambda numero: math.log10(numero)
resultado = logaritmo(n1)
print(f"The Log10 is: {resultado}")

seno = lambda numero: math.sin(math.radians(numero))
resultado = seno(n1)
print(f"The Sin of {n1} is: {resultado}")

cos = lambda numero: math.cos(math.radians(numero))
resultado = cos(n1)
print(f"The Cos of {n1} is: {resultado}")

tan = lambda numero: math.tan(math.radians(numero))
resultado = tan(n1)
print(f"The Tan of {n1} is: {resultado}")
```

```python
def factorial(n):

    if n == 0:  
        return 1
    else:  
        return n * factorial(n-1)

n = 3
resultado = factorial(n)
print(f"El factorial de {n} es {resultado}")
```

```python
def factorial2(n):
    if n < 0:
        return "Solo numeros positivos"
    elif n == 0:
        return 1
    else:
        val_fact = 1
        for i in range(1, n + 1):
            val_fact *= i
        return val_fact

num = 3
resultado = factorial2(num)
print(f"Factorial de {num} es {resultado}")
```

```python
def fibonacci(n): # n debe ser un numero entero positivo incluyendo el cero (numero natural)
     if n <= 1:
         return n
     else:
         return fibonacci(n - 1) + fibonacci(n - 2) # aquí se aplica la recursividad pues dentro del mismo bloque de código llamo a la función

#numero_de_terminos = 10
for i in range(21):
  print(fibonacci(i))
```

```python
def menu():
    while True:
        print("\n=== MENÚ PRINCIPAL ===")
        print("1. Saludar")
        print("2. Otra opción")
        print("3. Salir")

        opcion = input("Selecciona una opción: ")

        if opcion == "1":
            print("¡Hola!")
        elif opcion == "2":
            print("Opción en desarrollo...")
        elif opcion == "3":
            print("Saliendo del programa...")
            return  # Sale de la función y termina el bucle
        else:
            print("Opción no válida. Intenta nuevamente.")

menu()
```

```python
import unicodedata

def quitar_tildes(texto):
    """
    Quita las tildes de una cadena de texto.

    Args:
        texto: La cadena de texto de la que se quitarán las tildes.

    Returns:
        La cadena de texto sin tildes.
    """
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )

# Ejemplo de uso
texto_con_tildes = "áéíóúüÁÉÍÓÚÜñÑ"
texto_sin_tildes = quitar_tildes(texto_con_tildes)
print(f"Texto original: {texto_con_tildes}")
print(f"Texto sin tildes: {texto_sin_tildes}")
```
