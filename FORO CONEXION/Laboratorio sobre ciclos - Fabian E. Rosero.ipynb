  1. Solicite al usuario ingresar 5 nombres y guárdelos en una lista.
  
  2. Muestre cuántos nombres comienzan con la letra "A".

```python
nombres = []
num_elem = int(input("¿Cuántos nombres deseas ingresar? "))
for i in range(num_elem):
    elemento = input(f"Ingrese un nombre {i+1}: ")
    nombres.append(elemento)

print("La lista ingresada es:", nombres)

```

```python
nombres = []
num_elem = int(input("¿Cuántos nombres deseas ingresar? "))
cont = 0
for i in range(num_elem):
    elemento = input(f"Ingrese un nombre {i+1}: ")
    nombres.append(elemento)
    if "A" in elemento:
        cont += 1


print("La lista ingresada es:", nombres)
print ('La cantidad de nombres con letra A son:',cont)

```
