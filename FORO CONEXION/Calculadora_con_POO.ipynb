```python
import math # Importa el módulo math para funciones matemáticas como logaritmo, seno, coseno y tangente

class Calculadora: # Define la clase Calculadora para encapsular la funcionalidad
    def __init__(self): # Constructor de la clase Calculadora
        # Inicializa los atributos para almacenar los números de la operación
        self.numero_uno = None
        self.numero_dos = None

    def mostrar_titulo(self): # Método para mostrar el título de la calculadora
        print('Calculadora Basica COn Funciones usando (not) e (in) y (None)')

    def mostrar_menu(self): # Método para mostrar el menú de opciones al usuario
        print()
        print("MENU:")
        print("1: Sumar (+)")
        print("2: Restar (-)")
        print("3: Multiplicar (*)")
        print("4: Dividir (/)")
        print("5: Potencia (pot)")
        print("6: Radicación (rad)")
        print("7: Logaritmo (log10)")
        print("8: Seno (sen)")
        print("9: Coseno (cos)")
        print("10: Tangente (tan)")
        print("11: Salir")
        print()

    def ingresar_datos(self): # Método para solicitar al usuario que ingrese la opción y los números
        # Solicita al usuario que ingrese una opción del menú y los números necesarios para la operación
        opcion = input("Ingrese la opción deseada del menú: ").lower() # Convertir a minúsculas para manejar entradas como 'Sumar', 'SUMAR', etc.

        # Verifica si la opción requiere dos números
        if opcion in ['1','+','sumar',
                      '2','-','resta',
                      '3','*','multiplicacion',
                      '4','/','dividir',
                      '5','pot','potencia',
                      '6','rad','radicacion']:
            # Para las operaciones que requieren dos números
            try: # Manejo de errores para la entrada de números
                self.numero_uno = float(input('Ingrese el primer número de la operación: ')) # Almacena el primer número en el atributo de instancia
                self.numero_dos = float(input('Ingrese el segundo número de la operación: ')) # Almacena el segundo número en el atributo de instancia
                # Retorna la opción a la función que la procesará
                return opcion
            except ValueError: # Si la entrada no es un número válido
                print("Entrada inválida. Por favor ingrese números.")
                self.numero_uno = None # Restablece los atributos si hay un error
                self.numero_dos = None
                return None # Retorna None para indicar un error
        ## Para las operaciones que requieren un solo número (ángulo o valor)
        elif opcion in ['7', '8', '9', '10', 'log10', 'logaritmo', 'sen', 'cos', 'tan']: # Incluye las opciones de un solo número
            try: # Manejo de errores para la entrada de números
                self.numero_uno = float(input('Ingrese un número (ángulo o valor): ')) # Almacena el número en el atributo de instancia
                self.numero_dos = None # Asegura que el segundo número sea None para operaciones unitarias
                return opcion
            except ValueError: # Si la entrada no es un número válido
                print("Entrada inválida. Por favor ingrese un número.")
                self.numero_uno = None # Restablece el atributo si hay un error
                return None # Retorna None para indicar un error
        else:
            self.numero_uno = None # Restablece los atributos para opciones no válidas o 'salir'
            self.numero_dos = None
            return opcion  # Retorna la opción (incluyendo '11' para salir)

    # FUNCIONES MATEMÁTICAS (Métodos de la clase que usan los atributos de instancia)
    def calcular_suma(self): return self.numero_uno + self.numero_dos # funcion para sumar los números almacenados
    def calcular_resta(self): return self.numero_uno - self.numero_dos # funcion para restar los números almacenados
    def calcular_multiplicacion(self): return self.numero_uno * self.numero_dos # funcion para multiplicar los números almacenados
    def calcular_division(self): return self.numero_uno / self.numero_dos if self.numero_dos != 0 else None  # funcion para dividir los números almacenados, manejando división por cero
    def calcular_pot(self): return self.numero_uno ** self.numero_dos # funcion para elevar el primer número al segundo número almacenado
    def calcular_rad(self): # funcion para calcular la raíz n-ésima de los números almacenados
         if self.numero_uno < 0 and self.numero_dos % 2 == 0: # Maneja raíces pares de números negativos
             return None # Retorna None si es una raíz par de un número negativo
         return pow(self.numero_uno, 1/self.numero_dos)

    def calcular_logaritmo_base10(self): # funcion para calcular el logaritmo base 10 del número almacenado
        if self.numero_uno <= 0: # Maneja logaritmo de números no positivos
            return None # Retorna None si el número no es positivo
        return math.log10(self.numero_uno)
    def calcular_sen(self): return math.sin(math.radians(self.numero_uno)) # funcion para calcular el seno del ángulo almacenado en grados
    def calcular_cos(self): return math.cos(math.radians(self.numero_uno)) # funcion para calcular el coseno del ángulo almacenado en grados
    def calcular_tan(self): # funcion para calcular la tangente del ángulo almacenado en grados
        # Maneja casos donde la tangente es indefinida (múltiplos de 90 + 180*n)
        if math.cos(math.radians(self.numero_uno)) == 0:
            return None # Retorna None si la tangente es indefinida
        return math.tan(math.radians(self.numero_uno))

    ## Procesar la opción seleccionada y mostrar el resultado (Método de la clase)
    def procesar_y_mostrar_operacion(self, opcion): # Ahora solo necesita la opción ya que los números están en los atributos
        if opcion in ['1', '+', 'sumar']:
            print("La suma es:", self.calcular_suma()) # se llama al método de suma sin pasar números, usa los atributos de instancia
        elif opcion in ['2', '-', 'restar']:
            print("La resta es:", self.calcular_resta())
        elif opcion in ['3', '*', 'multiplicar']:
            print("La multiplicación es:", self.calcular_multiplicacion())
        elif opcion in ['4', '/', 'dividir']:
            resultado = self.calcular_division()
            if resultado is not None: # Verifica si la división fue posible
                print("La división es:", resultado)
            else:
                print("¡No se puede dividir entre cero!")
        elif opcion in ['5', 'pot', 'potencia']:
            print(f"{self.numero_uno} elevado a {self.numero_dos} es:", self.calcular_pot())
        elif opcion in ['6', 'rad', 'radicacion']:
            resultado = self.calcular_rad()
            if resultado is not None: # Verifica si la radicación fue posible
                 print(f"La raíz {self.numero_dos} de {self.numero_uno} es:", resultado)
            else:
                 print("¡No se puede calcular la raíz par de un número negativo!")
        elif opcion in ['7', 'log10', 'logaritmo']:
            resultado = self.calcular_logaritmo_base10()
            if resultado is not None: # Verifica si el logaritmo fue posible
                print("Logaritmo base 10 de", self.numero_uno, "es:", resultado)
            else:
                print("¡No se puede calcular el logaritmo de un número no positivo!")
        elif opcion in ['8', 'sen']:
            print(f"Seno de {self.numero_uno} grados:", self.calcular_sen())
        elif opcion in ['9', 'cos']:
            print(f"Coseno de {self.numero_uno} grados:", self.calcular_cos())
        elif opcion in ['10', 'tan']:
            resultado = self.calcular_tan()
            if resultado is not None: # Verifica si la tangente fue posible
                 print(f"Tangente de {self.numero_uno} grados:", resultado)
            else:
                 print("¡La tangente para este ángulo es indefinida!")
        else: # Maneja opciones que no están en la lista de operaciones (aunque este caso debería manejarse antes)
            print("Opción no válida. Intenta nuevamente.")


    def ejecutar(self): # Método principal para ejecutar la calculadora
        self.mostrar_titulo() # Llama al método para mostrar el título

        while True: # Bucle principal para mantener la calculadora en funcionamiento
            self.mostrar_menu() # Llama al método para mostrar el menú
            opcion = self.ingresar_datos() # Llama al método para ingresar datos

            # Valida Si la opción es '11', imprime un mensaje de despedida y termina el bucle mediante el break
            if opcion == '11':
                print("¡Gracias por usar la calculadora! Hasta pronto.")
                break
            # valida si la opción es None (opción inválida durante la entrada de datos)
            elif opcion is None:
                 # El mensaje de error ya se mostró en ingresar_datos, simplemente continuamos el bucle
                 pass
            # valida si la opcion que se digitó por teclado no está en las opciones válidas dentro de la lista
            # puede ser un número o una cadena de texto que se digita por teclado si no se encuentra en la lista, imprime un mensaje de error
            elif opcion not in ['1','+','sumar',
                                '2','-','resta',
                                '3','*','multiplicacion',
                                '4','/','dividir',
                                '5','pot','potencia',
                                '6','rad','radicacion',
                                '7','log10','logaritmo',
                                '8','sen',
                                '9','cos',
                                '10','tan',
                                '11']: # Incluye '11' en las opciones válidas para evitar el mensaje de error aquí
                print("Opción no válida. Intenta nuevamente.")
            else:
                self.procesar_y_mostrar_operacion(opcion) # envia la opción a la función procesar_y_mostrar_operacion para que realice la operación y muestre el resultado, usando los atributos de instancia

# EJECUCIÓN DEL PROGRAMA
if __name__ == "__main__": # Asegura que el código se ejecute solo cuando el script es corrido directamente
    calculadora = Calculadora() # Crea una instancia de la clase Calculadora
    calculadora.ejecutar() # Llama al método ejecutar para iniciar la calculadora
```
